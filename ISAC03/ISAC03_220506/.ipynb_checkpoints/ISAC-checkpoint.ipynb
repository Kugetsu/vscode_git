```python
list_kx=[] #グラフのx軸
list_ky=[]
list_kz=[]
list_eigen=[[[0] * 21 for i in range(21)] for j in range(21)] #固有値
```

```python
#K0-12にK14-26の転置を代入
for i in range(0,13):
    exec("matK"+str(i)+"=matK"+str(26-i)+".T")
#matkの代入
for a in range(0,3)
    exec("matk"+str(a*9+1)+"=matk"+str(a*9+4)+".T")
for i in range(0,8)
    exec("matk"+str(i*3)+"=matk"+str(22-i*3)+"")

#フォノンバンドのk点の個数
kx_num=10
ky_num=10
kz_num=10

for i in range(0,27):
    exec("matD=np.dot(matG,matK"+str(i)+")")
    exec("matK"+str(i)+"=np.dot(matD,matG)")
    exec("matD=np.dot(matG,matk"+str(i)+")")
    exec("matk"+str(i)+"=np.dot(matD,matG)")

#2020/11/19 15:13

for kx in range(-kx_num, kx_num+1):
    list_kx.append(kx/kx_num)
    for ky in range(-ky_num, ky_num+1):
        for kz in range(-kz_num, kz_num+1):
            a_i=(pi*kx/kx_num)*i_num
            b_i=(pi*ky/ky_num)*i_num
            c_i=(pi*kz/kz_num)*i_num
            
            matK=np.zeros((base_num,base_num),dtype=np.complex)
            
            
            for a in range(-1,2):
                for b in range(-1,2):
                    for c in range(-1,2):
                        exec("matK+=matK"+str((a+1)*9+(b+1)*3+c+1)+"*cmath.exp(a*a_i+b*b_i+c*c_i)")
            
            if(kx==0):
                if(ky==0):
                    if(kz==0):
                        #print(str(kx)+", "+str(ky)+", "+str(kz))
                        print(matK)
            '''
            matD[0,3]=0
            matD[1,3]=0
            matD[2,3]=0
            matD[3,0]=0
            matD[3,1]=0
            matD[3,2]=0
            '''
            
            l,P=np.linalg.eig(matK)

            l=np.sort(l,axis=0)
            v=np.transpose(P)

            list_eigen[kx+kx_num][ky+ky_num][kz+kz_num]=l.tolist()

for ky in range(-ky_num, ky_num+1):
    list_ky.append(ky/ky_num)

for kz in range(-kz_num, kz_num+1):
    list_kz.append(kz/kz_num)
```

```python
#kx方向 ykx[kyの値][kzの値][固有値の番号]
ykx=[[[[0.0+0j]*21 for k in range(base_num)] for i in range(21)] for j in range(21)]
for ky in range(0,21):
    for kz in range(0,21):
        for value in range(0,base_num):
            for i in range(0,21):
                ykx[ky][kz][value][i]=list_eigen[i][ky][kz][value]
            
#ky方向 yky[kxの値][kzの値][固有値の番号]
yky=[[[[0.0+0j]*21 for k in range(base_num)] for i in range(21)] for j in range(21)]
for kx in range(0,21):
    for kz in range(0,21):
        for value in range(0,base_num):
            for i in range(0,21):
                yky[kx][kz][value][i]=list_eigen[kx][i][kz][value]

#kz方向 ykz
ykz=[[[[0.0+0j]*21 for k in range(base_num)] for i in range(21)] for j in range(21)]
for kx in range(0,21):
    for ky in range(0,21):
        for value in range(0,base_num):
            for i in range(0,21):
                ykz[kx][ky][value][i]=list_eigen[kx][ky][i][value]
```
